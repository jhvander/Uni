# IS-LM Model Stability Analysis of the Equilibrium Point for a Closed Economy

In this project, we will model the stability of the investment-savings (IS) liquidity preference-money supply (LM) model, which is inspired by, $\textit{Nascimento, A. et.al. (2019)}$. They find that the equilibrium point of the dynamic IS-LM model is asymptotically stable for any variation in its parameters. Therefore, we will try to replicate their results and assess whether or not we can confirm their findings. 

The model is a Keynesian macroeconomic model that illustrates how the goods and the money market interact with each other to find an equilibrium between the interest rates and the output level. It is a simple macroeconomic model which allows policymakers to analyze the economy by simulating changes in different variables. In this study, we assume a closed economy, which implies that countries are self-sufficient and autarkic and that the interest rate is controlled by the equilibrium of supply and demand for money. Furthermore, we consider the private sector, the government, and a central bank.
 We define the model as follows:
\begin{equation*}
Y(t)=C(t)+I(t)+G(t),
\end{equation*}
where $Y(t)$ is the gross domestic product at time t, $C(t)$ is the private consumption, $I(t)$ is the investment and $G(t)$ is the government spending. Private consumption is defined as:
\begin{equation*}
C(t)=a+bY_d(t),
\end{equation*}
where $a$ is the autonomous consumption, $b$ is the marginal propensity to consume, $Y_d$ is the disposable income, which is defined as:
\begin{equation*}
Y_d(t)=Y(t)-T(t),
\end{equation*}
where $Y(t)$ is referred as the GDP (or income) and $T(t)$ is the tax defined as:
\begin{equation*}
T(t)=T_0+\lambda Y(t),
\end{equation*}
where $T$ is the fixed tax rate and $\lambda Y(t)$ measures the sensitivity of the tax rate to the GDP. Next, we define the investment as:
\begin{equation*}
I(t)=I_0-h r(t),
\end{equation*}
where $I_0$ is the autonomous investment, $h$ is the sensitivity of the investment to the interest rate $r(t)$. 
Lastly, we set the government to be exogenous.
\begin{equation*}
G(t)=\bar{G}.
\end{equation*}

We now define the money market, where the demand for money is defined as:
\begin{equation*}
M_d(t)=M_0+ kY(t)-\mu r(t),
\end{equation*}
where $M_0$ is the autonomous demand for money, $k$ is the sensitivity of the demand for money to the GDP, $\mu r(t)$ is the sensitivity of the demand for money to the interest rate $r(t)$ and reduces the demand for money as the interest rate increases. 
The supply of money is exogenous and defined as:
\begin{equation*}
M_s(t)=\bar{M}.
\end{equation*}
The market clearing condition is defined as:
\begin{equation*}
M_d(t)=M_s(t),
\end{equation*}
which implies that the demand for money equals the supply for money.

In this project, we will model the change in output concerning time t by taking the derivative of the GDP to time t:
\begin{equation*}
\frac{dy}{dt}=\alpha(C(t)+I(t)+G(t)-Y(t)),
\end{equation*}
where $\alpha$ is the speed of adjustment to the equilibrium.

We will now plug in the definitions for the variables:
\begin{equation*}
\frac{dy}{dt}=\alpha(a+b(Y(t)-T(t))+I_0-h r(t)+\bar{G}-Y(t)),
\end{equation*}
which can be rewritten as:
\begin{equation*}
\frac{dy}{dt}=\alpha(a+b(Y(t)-(T_{0}+\lambda Y(t)))+I_{0}-h r(t)+\bar{G}-Y(t))
\end{equation*}
where,
\begin{equation*}
\frac{dy}{dt}=\alpha(a-bT_{0}+I_{0}+\bar{G})-\alpha((1-b)(1-\lambda)Y(t))-\alpha h r(t).
\end{equation*}

We will do the same for the market clearing condition, where we want to formulate a differential equation for the interest rate w.r.t. time t:
\begin{equation*}
\frac{dr}{dt}=\beta(M_{d}(t)-M_{s}(t))
\end{equation*}
where $\beta$ is the speed of adjustment to the equilibrium in the money market.
We assume that $\beta>\alpha$. Meaning that the equilibrium interest rate adjusts faster than the equilibrium output. We will now plug in the definitions for the variables: 
\begin{equation*}
\frac{dr}{dt}=\beta(M_{0}+kY(t)-\mu r(t)-\bar{M}).
\end{equation*}

In this project, we will model the differential equations, $\frac{dy}{dt}$, and $\frac{dr}{dt}$, with $Y(t)$ and $r(t)$ being the endogenous variables in the model. Whereas, $T_{0}$, $I_{0}$, $M_{0}$, are the exogenous variables in the model. These variables will be adjusted by policymakers to simulate different scenarios of the economy. 

# ODE solver method

ODE solver methods encompass a range of numerical techniques employed to approximate solutions to ordinary differential equations. These equations describe the relationship between a function and its derivatives. We present the key ODE solver methods, their relevant equations, and properties as follows:

- **Euler's method** is a first-order numerical technique for solving ODEs. It employs a simple, iterative approach that uses a finite step size to approximate the solution at subsequent points. Given an initial value problem:

\begin{equation*}
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
\end{equation*}

The Euler's method update formula is:

\begin{equation*}
y(t + h) = y(t) + h \cdot f(t, y(t))
\end{equation*}

where *h* is the step size, and *t* is incremented by *h* in each iteration.

- **Runge-Kutta methods** offer higher-order, more accurate numerical techniques for solving ODEs. The most widely used is the fourth-order Runge-Kutta method (RK4), which employs intermediate steps to improve the accuracy of the solution. Given the same initial value problem as before, the RK4 method involves the following steps:

\begin{align}
k_1 &= h \cdot f(t, y(t)) \\
k_2 &= h \cdot f(t + \frac{h}{2}, y(t) + \frac{k_1}{2}) \\
k_3 &= h \cdot f(t + \frac{h}{2}, y(t) + \frac{k_2}{2}) \\
k_4 &= h \cdot f(t + h, y(t) + k_3)
\end{align}

\begin{equation*}
y(t + h) = y(t) + \frac{k_1 + 2k_2 + 2k_3 + k_4}{6}
\end{equation*}

- **Multistep methods**, such as the Adams-Bashforth and Adams-Moulton families, utilize information from previous steps to achieve higher accuracy in solving ODEs. These methods rely on both implicit and explicit formulas to approximate the solution at future points. A popular choice is the Predictor-Corrector method, which combines the explicit Adams-Bashforth method as the predictor and the implicit Adams-Moulton method as the corrector.

- **Stiff ODEs** exhibit rapid changes in their solutions, necessitating specialized solver methods to maintain accuracy and stability. Examples of such solvers include backward differentiation formulas (BDF) and the Rosenbrock methods. These methods are implicit in nature and typically require solving nonlinear systems of equations at each step.

In summary, ODE solver methods are vital tools for approximating the solutions to ordinary differential equations in various fields. The choice of an appropriate method depends on the problem's characteristics, such as desired accuracy, stability, and computational efficiency. The paper is based on the Euler's method and the RK method.

## Compare ODE solver, RK4 and Forward Euler
First, we will compare different solver methods and asses how the results might differ according to, $\textit{Nascimento, A. et.al. (2019)}$. We will solve for the differential equations derived above for the interest rate (r) and GDP (y).

```python

# Define parameters

a = 15 # autonomous consumption
I_0 = 10 # autonomous investment
T_0 = 0 # zero fixed taxes
M_0 = 0 # zero autonomous demand for money
b = 0.75 # marginal propensity to consume 
h = 1.525 # sensitivity of investment in relation to interest rates
k = 0.25 # sensitivity of demand for money in relation to GDP
mu_val = 0.5 # sensitivity of demand for money in relation to interest rates
G = 25 # government spending is exogenous
M = 8 # money supply is exogenous
alpha_val = 0.05 # speed of adjustment in the goods market
beta_val = 0.8  # speed of adjustment in money market
lambda_val = 0.25 # sensitivity of taxes in relation to GDP


time_points = np.arange(0, 100) # Define time horizon for simulation

#Define the IS-LM model differential equations
def IS_LM_model(u, t):    
    GDP, interest_rate = u
    dYdt = ((alpha_val * (a - b * T_0 + I_0 + G)) - 
            (alpha_val * ((1-b) * (1-lambda_val) * GDP)) - 
            alpha_val * h * interest_rate) 
    drdt = beta_val * (M_0 - M + k * GDP - (mu_val * interest_rate))
    return [dYdt, drdt]


# Initial conditions vector
Y_0 = 75
r_0 = 8

# RK4 solver
solver_RK4 = ISLM.RungeKutta4(IS_LM_model) # import the solver method from ISLM.py
solver_RK4.set_initial_condition([Y_0, r_0]) # setting the initial conditions.
u, t = solver_RK4.solve(time_points) # solve with the timepoints defined above
GDP_RK4 = u[:,0] # GDP being equal to all the values stored in the first column of u
interest_rate_RK4 = u[:,1]

# Forward Euler solver method

# Import the ForwardEuler solver method from ISLM.py
solver_FE = ISLM.ForwardEuler(IS_LM_model) 

# Set the initial conditions for the solver
solver_FE.set_initial_condition([Y_0, r_0])

# Solve the system of differential equations using the ForwardEuler method
u, t = solver_FE.solve(time_points)

# Extract the GDP and interest rate solutions from the solver output
GDP_FE = u[:, 0]
interest_rate_FE = u[:, 1]


# The IS-LM model using the ODE function 
def system1(init_values, t, a, I_0, T_0, M_0, G, M, alpha_val, beta_val, 
           lambda_val, b, h, k, mu_val):
    GDP, interest_rate = init_values # Setting these equal to the inital values defined above.
    dYdt = (alpha_val * (a - b * T_0 + I_0 + G) - 
            alpha_val * ((1-b) * (1-lambda_val) * GDP) - 
            (alpha_val * h * interest_rate))  #defining the defferential equations.
    drdt = beta_val * (M_0 - M + k * GDP - (mu_val * interest_rate))
    return dYdt, drdt

init_conds = Y_0, r_0
args_tuple = (a, I_0, T_0, M_0, G, M, alpha_val, beta_val, lambda_val, b, h, k, mu_val)

# Using the odeint function from scipy
ret = odeint(system1, init_conds, time_points, args=args_tuple)

GDP_odeint, interest_rate_odeint = ret.T

#plots the function
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))
ax = ax.flatten()
ax[0].plot(t, GDP_RK4, 'b', alpha=0.4, lw=2, label='GDP_RK4')
ax[0].plot(t, GDP_odeint, 'r', alpha=0.4, lw=2, label='GDP_ode')
ax[0].plot(t, GDP_FE, 'g', alpha=0.4, lw=2, label='GDP_FE')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('GDP')
ax[0].set_title('GDP Solution Path', fontsize=13)
ax[0].legend(loc=4)
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)

ax[1].plot(t, interest_rate_RK4, 'b', alpha=0.4, lw=2, label='interest_rate_RK4')
ax[1].plot(t, interest_rate_odeint, 'r', alpha=0.4, lw=2, label='interest_rate_ode')
ax[1].plot(t, interest_rate_FE, 'g', alpha=0.4, lw=2, label='interest_rate_FE')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('interest rate')
ax[1].set_title('Interest Rate Solution Path', fontsize=13)
ax[1].legend(loc=4)
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
plt.show()
```

We find that the Forward Euler method generates higher estimates of the GDP and the interest rate values, but tends to move towards the equilibrium along with the two other methods. Furthermore, the results provided by RK4 and Scipy's ODE solver are very similar to $\textit{Nascimento, A. et.al. (2019)}$ findings. In fact, in their paper, they also use the ODE solver method. This is because the interest rate is the cost of borrowing money, and when the interest rate increases the cost of borrowing money increases, which leads to a decrease in investment and consumption. This is also the case when the interest rate decreases, and the cost of borrowing money decreases, which leads to an increase in investment and consumption. Therefore, we would expect that the instantaneous increase in the interest rate would consequently
cause a decrease in the GDP of the economy. However, if we adjust the starting position of GDP to 80, we would observe the same pattern as in $\textit{Nascimento, A. et.al. (2019)}$. This indicates that 75 is below the equilibrium for GDP, why we observe an increase in GDP when the interest rate increases. However, we still observe that both the interest rate and the GDP moves towards their equilibrium.

This could lead to an increase in inflation, making the price of goods and services more expensive and thereby decreasing GDP in the long run. Therefore, the government must be aware when increasing government spending.

## Extensions of the IS-LM model

Now we extend the IS-LM model by introducing inflation to the model. Inflation is the rate at which the general price level of goods and services in the economy is increasing over time. The introduction of inflation would help to capture the dynamics of the economy more accurately since inflation affects both the demand and supply sides of the economy.

To introduce inflation, we need to modify the consumption and investment equations to include inflation expectations. In particular, we assume that consumers and firms have some expectation of future inflation, denoted by $\pi(t+1)$. We can modify the consumption equation as follows:

\begin{equation*}
C(t) = a + b(Y(t)-T(t)) - \sigma\pi(t+1),
\end{equation*}

where $\sigma$ is the sensitivity of consumption to expected inflation. The term $-\sigma\pi(t+1)$ represents the effect of expected inflation on consumption, which reduces consumption as expected inflation increases.

Similarly, we can modify the investment equation as follows:

\begin{equation*}
I(t) = I_0 - hr(t) - \gamma\pi(t+1),
\end{equation*}

where $\gamma$ is the sensitivity of investment to expected inflation. The term $-\gamma\pi(t+1)$ represents the effect of expected inflation on investment, which reduces investment as expected inflation increases.

We also need to modify the money market equation to include expected inflation:

\begin{equation*}
M_d(t) = M_0 + kY(t) - \mu r(t) + \theta\pi(t+1),
\end{equation*}

where $\theta$ is the sensitivity of the demand for money to expected inflation. The term $\theta\pi(t+1)$ represents the effect of expected inflation on the demand for money, which increases the demand for money as expected inflation increases.

Lastly, we need to introduce an equation for inflation dynamics. A simple way to model inflation is to use the Phillips curve, which relates inflation to the output gap (the difference between actual output and potential output). The Phillips curve equation is:

\begin{equation*}
\pi(t) = \phi(Y(t)-\bar{Y}),
\end{equation*}

where $\phi$ is the sensitivity of inflation to the output gap, and $\bar{Y}$ is the potential output of the economy.

With these modifications, the extended IS-LM model becomes:

\begin{align}
Y(t) &= C(t) + I(t) + G(t), \\
C(t) &= a + b(Y(t)-T(t)) - \sigma\pi(t+1), \\
I(t) &= I_0 - hr(t) - \gamma\pi(t+1), \\
Y_d &= Y(t)-T(t) \\
M_d(t) &= M_0 + kY(t) - \mu r(t) + \theta\pi(t+1), \\
M_s(t) &= \bar{M}, \\
M_d(t) &= M_s(t), \\
\pi(t) &= \phi(Y(t)-\bar{Y}).
\end{align}

This extended model can be used to analyze the effects of inflation expectations on the behaviour of the economy. For example, an increase in expected inflation would reduce consumption and investment, as well as increase the demand for money. This in turn would lead to a decrease in output and an increase in the interest rate.

We now derive the extended model
\begin{equation*}
\frac{dy}{dt}=\alpha(C(t)+I(t)+G(t)-Y(t)),
\end{equation*}
where $\alpha$ is the speed of adjustment to the equilibrium.

We will now plug in the definitions for the variables:
\begin{equation*}
\frac{dy}{dt}=\alpha(a+b(Y(t)-T(t))+I_0-h r(t)+\bar{G}-Y(t)),
\end{equation*}
which can be rewritten as:
\begin{equation*}
\frac{dy}{dt}=\alpha(a+b(Y(t)-(T_{0}+\lambda Y(t))-\sigma\pi(t+1))+I_{0}-hr(t)-\gamma\pi(t+1)+\bar{G}-Y(t))
\end{equation*}
where,
\begin{equation*}
\frac{dy}{dt}=\alpha(a-bT_{0}+I_{0}+\bar{G}-b\sigma\pi(t+1))-\alpha((1-b)(1-\lambda)Y(t))-\alpha hr(t)-\gamma\pi(t+1).
\end{equation*}

We will do the same for the market clearing condition, where we want to formulate a differential equation for the interest rate w.r.t. time t:
\begin{equation*}
\frac{dr}{dt}=\beta(M_{d}(t)-M_{s}(t))
\end{equation*}
where $\beta$ is the speed of adjustment to the equilibrium in the money market.
We assume that $\beta>\alpha$. Meaning that the equilibrium interest rate adjusts faster than the equilibrium output. We will now plug in the definitions for the variables: 
\begin{equation*}
\frac{dr}{dt}=\beta(M_{0}+kY(t)-\mu r(t)+\eta\pi(t+1)-\bar{M}).
\end{equation*}


As in the base case, we set the potential output $\bar{Y}=Y_0$ being 75. Next, we set the paramters $\theta=0.3, \sigma=0.1, \gamma=0.1$ and $\eta=0.1$. The rest of the parameters are equal to the ones in the base case. Now, we will compare how the extended model behaves compared to the base case.

```python

#### list of all parameters in the model

a = 15 # autonomous consumption
I_0 = 10 # autonomous investment
T_0 = 0 # no fixed taxes
M_0 = 0 # no autonomous demand for money
alpha_val = 0.05 # speed of adjustment in goods market
beta_val = 0.8  # speed of adjustment in money market
lambda_val = 0.25 # sensitivity of taxes in relation to GDP
b = 0.75 # marginal propensity to consume 
h = 1.525 # sensitivity of investment in relation to interest rates
k = 0.25 # sensitivity of demand for money in relation to GDP
mu_val = 0.5 # sensitivity of demand for money in relation to interest rates
G = 25 # government spending, exogenous
M = 8 # money supply, exogenous

theta = 0.3
sigma = 0.1
gamma = 0.1
eta = 0.1


time_points = np.arange(0, 100)


def IS_LM_model(u, t):    
    GDP, interest_rate = u
    dYdt = ((alpha_val * (a - b * T_0 + I_0 + G-sigma*theta*(GDP-Y_0)*GDP)) - 
            (alpha_val * ((1-b) * (1-lambda_val) * GDP)) - 
            alpha_val * h * interest_rate-gamma*theta*(GDP-Y_0)) 
    drdt = beta_val * (M_0 - M + k * GDP - (mu_val * interest_rate)+eta*theta*(GDP-Y_0)*M_0)
    return [dYdt, drdt]


# Initial conditions vector
Y_0 = 75
r_0 = 8
#


# The IS-LM model differential equations
def system(init_values, t, a, I_0, T_0, M_0, G, M, alpha_val, beta_val, 
           lambda_val, b, h, k, mu_val, theta, sigma, gamma, eta):
    GDP, interest_rate = init_values
    dYdt = ((alpha_val * (a - b * T_0 + I_0 + G-sigma*theta*(GDP-Y_0))) - 
            (alpha_val * ((1-b) * (1-lambda_val) * GDP)) - 
            alpha_val * h * interest_rate-gamma*theta*(GDP-Y_0)) 
    drdt = beta_val * (M_0 - M + k * GDP - (mu_val * interest_rate)+eta*theta*(GDP-Y_0))
    return [dYdt, drdt]

init_conds = Y_0, r_0
args_tuple = (a, I_0, T_0, M_0, G, M, alpha_val, beta_val, lambda_val, b, h, k, mu_val, theta, sigma, gamma, eta)

# Integrate the IS-LM differential equations over the time grid, t.
ret = odeint(system, init_conds, time_points, args=args_tuple)

GDP_odeint_inflation, interest_rate_odeint_inflation = ret.T

```

```python
plot_inflation(GDP_odeint_inflation, interest_rate_odeint_inflation, time_points, 'Expansionary Fiscal Policy in a Pandemic')
```

In the right-hand side of the figure above, we observe that the equilibrium of the GDP with inflation is now lower compared to the base case. The reason for this is due to inflation expectations. As inflation expectations increase, consumption and investment decrease which would lead to a decrease in output, but at the same time increase in interest rate. However, we only observe a short increase in the interest rate, when the te output gap is positive. When GDP decreases the output gap moves to 0, meaning that inflation expectations have a smaller effect on the interest rate and that is why we do not see the interest rate being above the base interest rate. This contradicts the theory behind the implemented inflation expectations, and we would instead have expected to see the interest rate with inflation being above the base interest rate. But, we do observe that the interest with inflation has an overall increase, which is in line with the theory.
